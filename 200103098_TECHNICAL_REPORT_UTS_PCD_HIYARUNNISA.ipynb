# Soal 2
Gunakan konsep konvolusi untuk melakukan smoothing pada citra berwarna. Lakukan langkah-langkah berikut:

*   Baca citra "namalengkap.jpg" menggunakan library OpenCV (cv).
*   Terapkan konvolusi antara citra dan kernel dengan ukuran kernel 3x3 (terserah untuk kernel yang digunakan).
*   Tampilkan citra asli, identity, penajaman, blurring, dan hasil averaging pixe


```python
# Pastikan citra telah terbaca dengan benar
if img is None:
    print("Gagal membaca citra.")
else:
    # Buat kernel untuk identity
    kernel_identity = np.array([[0, 0, 0],
                                [0, 1, 0],
                                [0, 0, 0]], np.float32)

    # Buat kernel untuk sharpening
    kernel_sharpening = np.array([[0, -1, 0],
                                  [-1, 5, -1],
                                  [0, -1, 0]], np.float32)

    # Buat kernel untuk blurring
    kernel_blurring = np.ones((5,5), np.float32) / 25

    # Buat kernel untuk averaging pixel
    kernel_averaging =  np.array([[1, 1, 1],
                                  [1, 1, 1],
                                  [1, 1, 1]], np.float32) / 9
```

1. Konvolusi Citra

Suatu citra digital didefiniskan sebagai  suatu fungsi f(x,y) terdiri  dari sejumlah baris dan  sejumlah kolom dengan x dan y adalah koordinat spasial, dan amplitude f di titik koordinat (x, y) dinamakan intensitas dari citra pada titik tersebut [1] [3]. Konvolusi dilakukan dengan mengalikan elemen-elemen pada dua buah matriks dan menjumlahkan hasilnya, sehingga terdapat keluaran nilai dari sebuah piksel yang merupakan hasil konvolusi.  Satu  dari  matriks  tersebut  merepresentasikan  citra  dan  yang  lain  merepresentasikan filter/tapis/kernel [4].

2. Pada penajaman citra,

Intensitas suatu citra akan diperkuat dengan melewatkan pada suatu tapis lolos atas (high pass filter).  Ada berbagai  macam kernel  konvolusi yang  membentuk high-pass filter,  ada yang menonjolkan arah-arah tertentu (vertikal, horisontal, diagonal, atau juga utara, selatan, barat dan timur) [4]. Penajaman citra bisa terjadi karena citra tersebut didiferensiasi (dikurangi) dengan matriks kernel (h). Hasil penjumlahan tiap elemen pada matriks kernel adalah 1.

```python
# Terapkan konvolusi antara citra dan kernel untuk identity
img_identity = cv2.filter2D(img, -1, kernel_identity)

# Terapkan konvolusi antara citra dan kernel untuk penajaman
img_sharpening = cv2.filter2D(img, -1, kernel_sharpening)

# Terapkan konvolusi antara citra dan kernel untuk blurring
img_blurring = cv2.filter2D(img, -1, kernel_blurring)

# Terapkan konvolusi antara citra dan kernel untuk averaging pixel
img_averaging = cv2.filter2D(img, -1, kernel_averaging)
```

```python

# Tampilkan citra asli, hasil smoothing, hasil penajaman, hasil blurring, dan hasil averaging pixel
plt.figure(figsize=(20, 10))

plt.subplot(2, 4, 1)  # Membuat subplot pada baris pertama, kolom pertama
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # Menampilkan citra asli
plt.title('Citra Asli')  # Menambahkan judul 'Citra Asli'
plt.axis('off')  # Matikan sumbu

plt.subplot(2, 4, 2)  # Membuat subplot pada baris pertama, kolom kedua
plt.imshow(cv2.cvtColor(img_sharpening, cv2.COLOR_BGR2RGB))  # Menampilkan citra setelah proses sharpening
plt.title('Sharpening')  # Menambahkan judul 'Sharpening'
plt.axis('off')  # Matikan sumbu

plt.subplot(2, 4, 3)  # Membuat subplot pada baris pertama, kolom ketiga
plt.imshow(cv2.cvtColor(img_blurring, cv2.COLOR_BGR2RGB))  # Menampilkan citra setelah proses blurring
plt.title('Blurring')  # Menambahkan judul 'Blurring'
plt.axis('off')  # Matikan sumbu

plt.subplot(2, 4, 4)  # Membuat subplot pada baris pertama, kolom keempat
plt.imshow(cv2.cvtColor(img_averaging, cv2.COLOR_BGR2RGB))  # Menampilkan citra setelah proses averaging pixel
plt.title('Averaging Pixel')  # Menambahkan judul 'Averaging Pixel'
plt.axis('off')  # Matikan sumbu

plt.show()  # Menampilkan plot
```

### Segmentasi citra
Segmentasi  citra  adalah  sebuah  proses  untuk  memisahkan  sebuah  objek  dari  background, sehingga objek tersebut dapat diproses untuk keperluan lain. Dengan proses segmentasi tersebut, masing-masing  objek  pada  gambar  dapat  diambil  secara  individu  sehingga  dapat  digunakan sebagai  input  bagi  proses  yang  lain,  sebagai  contoh,  padaproses  rekontruksi  objek  3  dimensi, diperlukan   proses   segmentasi   untuk   memisahkan   objek   yang   akan   direkontruksi   terhadap background yang ada

```python
# a) Baca citra menggunakan OpenCV
gambar = cv2.imread('Foto Hia.jpg')
# b) Rentang warna grayscale yang ingin di-segmentasi
lower_gray = np.array([100], dtype=np.uint8)
upper_gray = np.array([200], dtype=np.uint8)
```

```python
# Segmentasi warna pada citra grayscale

#convert BGR ke RGB
gambar=cv2.cvtColor(gambar,cv2.COLOR_BGR2RGB)

#Konversi citra warna RGB menjadi citra grayscale
gray_gambar = cv2.cvtColor(gambar, cv2.COLOR_BGR2GRAY)

#Menciptakan mask berdasarkan rentang warna yang ditentukan pada citra grayscale
mask = cv2.inRange(gray_gambar, lower_gray, upper_gray)

#Menghasilkan citra hasil segmentasi dengan menerapkan masker yang telah dibuat sebelumnya ke citra asli
result = cv2.bitwise_and(gambar, gambar, mask=mask)
```

```python
# c) Tampilkan citra asli dan hasil segmentasi
plt.figure(figsize=(12, 6))

plt.subplot(1,3,1)
plt.imshow(gambar)
plt.title('Citra Asli')

plt.subplot(1,3,2)
plt.imshow(result, cmap= 'gray')
plt.title('Hasil Segmentasi Grayscale')

plt.show()
```

```python
# d) Tampilkan histogram warna
colors = ('b', 'g', 'r')
plt.figure(figsize=(12, 6))

# plot histogram untuk setiap channel warna
for i, col in enumerate(colors):  # Looping melalui daftar warna dengan indeks i dan nilai warna col
    hist = cv2.calcHist([gambar], [i], None, [256], [0, 256])  # Menghitung histogram untuk saluran warna i (R, G, atau B)
    plt.plot(hist, color=col)  # Plot histogram dengan warna yang sesuai
    plt.xlim([0, 256])  # Mengatur batas sumbu x dari 0 hingga 256 (rentang intensitas piksel)

plt.title('Histogram Warna')
plt.xlabel('Intensitas Piksel')
plt.ylabel('Jumlah Piksel')
plt.show()
```

# Soal 4
Lakukan equalisasi gambar

*   Baca citra "nim.jpg" menggunakan library OpenCV
*   Tentukan rentang warna grayscale yang ingin di-segmentasi
*   Tampilkan citra asli dan hasil segmentasi
*   Tampilkan hasil dari Histogram Gambar Asli, Equalisasi Histogram, dan brightened image histogram




Ekualisasi histogram adalah sebuah teknik pengolahan gambar yang digunakan untuk memperbaiki kontras dan warna gambar. Dalam proses ini, histogram gambar diubah agar memiliki distribusi yang lebih merata, sehingga warna dan kontras gambar menjadi lebih jelas dan lebih menarik. Teknik ini sangat berguna dalam pengolahan gambar yang memiliki kontras yang tidak merata, seperti gambar yang terlalu gelap atau terlalu terang.

```python
# a) Baca citra menggunakan OpenCV
gambar = cv2.imread('Foto Hia.jpg')

# b) Rentang warna grayscale yang ingin disegmentasi
lower_gray = np.array([100], dtype=np.uint8)
upper_gray = np.array([200], dtype=np.uint8)
```

```python
# Segmentasi warna pada citra grayscale

#convert BGR ke RGB
gambar=cv2.cvtColor(gambar,cv2.COLOR_BGR2RGB)

#Konversi citra warna RGB menjadi citra grayscale
gray_gambar = cv2.cvtColor(gambar, cv2.COLOR_BGR2GRAY)

#Menciptakan mask berdasarkan rentang warna yang ditentukan pada citra grayscale
mask = cv2.inRange(gray_gambar, lower_gray, upper_gray)

#Menghasilkan citra hasil segmentasi dengan menerapkan masker yang telah dibuat sebelumnya ke citra asli
result = cv2.bitwise_and(gambar, gambar, mask=mask)
```

```python
# c) Tampilkan citra asli dan hasil segmentasi
plt.figure(figsize=(12, 6))

plt.subplot(1,3,1)
plt.imshow(gambar)
plt.title('Citra Asli')

plt.subplot(1,3,2)
plt.imshow(result, cmap= 'gray')
plt.title('Hasil Segmentasi Grayscale')

plt.show()
```

```python

def histogram_equalization(image):
    # Mengkonversi citra menjadi grayscale jika belum dalam format tersebut
    if len(image.shape) == 3:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Menghitung histogram
    hist, bins = np.histogram(image.flatten(), 256, [0,256])

    # Menghitung kumulatif distribusi histogram
    cdf = hist.cumsum()
    cdf_normalized = cdf * hist.max() / cdf.max()

    # Membuat lookup table (LUT)
    lut = np.interp(image.flatten(), bins[:-1], cdf_normalized)

    # Menerapkan lookup table ke citra asli
    image_equalized = lut.reshape(image.shape)

    return image_equalized

# Memuat citra
image = cv2.imread('Foto Hia.jpg')

# Melakukan ekualisasi histogram
equalized_image = histogram_equalization(image)

# Menampilkan citra asli dan citra setelah ekualisasi histogram
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Citra Asli")
plt.subplot(1, 2, 2)
plt.imshow(equalized_image, cmap='gray')
plt.title("Citra Setelah Ekualisasi Histogram")
plt.show()
```

```python
# hasil dari Histogram Gambar Asli, Equalisasi Histogram, dan brightened image histogram

def equalize_histogram(image):
    # Menggunakan fungsi OpenCV untuk ekualisasi histogram
    equalized_image = cv2.equalizeHist(image)
    return equalized_image

def adjust_brightness(image, alpha):
    # Menggunakan operasi gamma correction untuk menyesuaikan kecerahan citra
    adjusted_image = cv2.convertScaleAbs(image, alpha=alpha, beta=0)
    return adjusted_image

def adjust_contrast(image, alpha, beta):
    # Menggunakan operasi linear transformation untuk menyesuaikan kontras citra
    adjusted_image = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)
    return adjusted_image

# Membaca citra grayscale
image = cv2.imread('Foto Hia.jpg', cv2.IMREAD_GRAYSCALE)

# Melakukan ekualisasi histogram
equalized_image = equalize_histogram(image)

# Menyesuaikan tingkat kecerahan
# High Brigthess
brightened_image_hi = adjust_brightness(image, alpha=1.5) # Ubah nilai alpha sesuai kebutuhan
# Low Brigthness
brightened_image_low = adjust_brightness(image, alpha=0.5)


# Menampilkan citra asli dan hasil-hasil pengolahan
plt.figure(figsize=(15, 8))

plt.subplot(4, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Gambar Original')
plt.subplot(4, 2, 2)
plt.hist(image.ravel(), 256, [0,256])
plt.title('Original Histogram')

plt.subplot(4, 2, 3)
plt.imshow(equalized_image, cmap='gray')
plt.title('Gambar yang sudah di Ekualisasi')
plt.subplot(4, 2, 4)
plt.hist(equalized_image.ravel(), 256, [0,256])
plt.title('Equalized Histogram')

plt.subplot(4, 2, 5)
plt.imshow(brightened_image_hi, cmap='gray')
plt.title('Gambar High Brightened')
plt.subplot(4, 2, 6)
plt.hist(brightened_image_hi.ravel(), 256, [0,256])
plt.title('High Brightened Histogram')

plt.subplot(4, 2, 7)
plt.imshow(brightened_image_low, cmap='gray')
plt.title('Gambar Low Brightened')
plt.subplot(4, 2, 8)
plt.hist(brightened_image_low.ravel(), 256, [0,256])
plt.title('Low Brightened Histogram')

plt.tight_layout()
plt.show()
```
